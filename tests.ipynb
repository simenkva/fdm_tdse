```python
import numpy as np
import matplotlib.pyplot as plt 
from icecream import ic
from fdm import fdm_laplacian_1d, fdm_laplacian_1d_nonuniform


delta, x = fdm_laplacian_1d(0, 10, 300, order=2)
x = x[1:-1]
delta = delta.todense()[1:-1, 1:-1]
#x = x[1:-1] + (x[1]-x[0])/2


H = -0.5*delta - np.diag(0.125/x**2) + np.diag(x**2/2)

omega, U = np.linalg.eigh(H)
ic(omega[:5])
ic(x.shape)
```

```python
plt.figure()
u = U[:,0].reshape((len(x),))
plt.plot(u / np.sqrt(x) , 'C0')
#plt.plot(x, np.exp(-x**2/2) * 0.1 , 'C0--')
# plt.plot(x, U[:, 1] / np.sqrt(x))
# plt.plot(x, U[:, 2] / np.sqrt(x))
plt.show()
```
